```python
from matplotlib import pyplot as plt

def comparison_plot(percentiles, num_bins, min_matches, min_matches_plot,
                    show_top_percentile, tol, mz_power, int_power, pick):
    #Compare all:
    plt.style.use('ggplot')
    fig, ax = plt.subplots(figsize=(7,6))

    x_percentiles = (show_top_percentile / num_bins * (1 + np.arange(num_bins)))[::-1]

    color_count = 0
    color_max = len(min_matches_plot) + 1
    for i, percentile in enumerate(percentiles):
        if min_matches[i] in min_matches_plot:
            plt.plot(x_percentiles, percentile, 
                     color=(color_count/color_max, 0, 1-color_count/color_max, 0.2+0.8*color_count/color_max),
                     label='{}) tol={}, mz_power={}, intensity_power={}, min_match>={}'.format((color_count+1), 
                                                                                               tol, mz_power, int_power, min_matches[i]))
            color_count += 1

    i = pick    
    plt.plot(x_percentiles, percentiles[i], 
             "--", color="black",
             label='selected Cosine score (tol={}, intensity_power={}, min_match>={})'.format(tol, int_power, min_matches[i]))

    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    plt.xticks(np.linspace(0, show_top_percentile, 5),
               ["{:.3f}%".format(x) for x in np.linspace(0, show_top_percentile, 5)])
    plt.legend(fontsize=8)
    plt.xlabel("Top percentile of spectral similarity score g(s1,s2)", fontsize=13)
    plt.ylabel("Mean molecular similarity f(m1,m2) \n (within respective percentile)", fontsize=13)
    plt.title("Cosine score vs. molecule fingerprint based similarity")
    plt.xlim(0, 0.1)
    plt.ylim(0.38, 0.92)
```

```python
num_bins = 100
show_top_percentile = 0.1
tol = 0.005
int_power = 1.0
mz_power = 0
min_matches = np.arange(1,21)
min_matches_plot = [1,2,4,6,8,10,15,20]
comparison_plot(percentiles_cosine, num_bins, min_matches, min_matches_plot,
                show_top_percentile, tol, mz_power, int_power, pick=5)
plt.savefig(os.path.join(path_data, 'Benchmarking_Cosine_tol0005_200708_min_matches.pdf'))
```
